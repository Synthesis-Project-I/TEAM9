# DATA CLEANING

```python
import pandas as pd
import numpy as np

file_path = "/content/drive/MyDrive/PROJECT SYNTHESIS I/260223_GrauIAI_data_mock.xlsx"    #with this we don't have to upload the data every time
df_dict = pd.read_excel(file_path, sheet_name=None)   #this creates a dictionary for our data where each key corresponds to a datasheet

df_dict.keys()
```

```python
#let's define a variable for each individual datasheet
df1 = df_dict['Data']
df2 = df_dict['Schedules']
df3 = df_dict['Clients']
df4 = df_dict['TranslatorsCost+Pairs']

df1.head()
```

```python
#we get rid of the columns we are not interested in
df1 = df1.drop(columns=['PROJECT_ID', 'PM', 'TASK_ID', "ASSIGNED", "RECEIVED", "CLOSE", "HOURS", "HOURLY_RATE","MANUFACTURER_SECTOR", "MANUFACTURER_INDUSTRY_GROUP",
                        "MANUFACTURER_INDUSTRY", "MANUFACTURER_SUBINDUSTRY"])
```

```python
df1.head()
```

```python
#we put the "TRANSLATOR" column in first position just porque sino me da toc
cols = ['TRANSLATOR'] + [c for c in df1.columns if c != 'TRANSLATOR']
df1 = df1[cols]

df1.head()
```

```python
#this creates a new column corresponding to the task EXPECTED length (END-START) in hours
df1['START'] = pd.to_datetime(df1['START'])
df1['END'] = pd.to_datetime(df1['END'])

df1['TASK_LENGTH'] = (df1['END'] - df1['START']).dt.total_seconds() / 3600

#this creates a new column corresponding to the amount of hours the translator spent working on the task (paid work), (DELIVERED-WORKING)
df1['WORKING'] = pd.to_datetime(df1['WORKING'])
df1['DELIVERED'] = pd.to_datetime(df1['DELIVERED'])

df1['TRANSLATOR_WORKING_TIME'] = (df1['DELIVERED'] - df1['WORKING']).dt.total_seconds() / 3600

df1.head(10)
```

```python
#we observe some inconsistencies. For example, that some task_length values are negative (shown below), some costs are 0, some TRANSLATOR_WORKING_TIME is
#too close to 0, or some source and target language being the same
df1[df1['TASK_LENGTH'] < 0]
```

```python
#let's clean the data
df1 = df1[(df1['TASK_LENGTH'] > 0.01) & (df1['TRANSLATOR_WORKING_TIME'] > 0.01) & (df1['SOURCE_LANG'] != df1['TARGET_LANG']) & (df1['COST'] > 0)]

df1.head(10)
```

```python
#finally let's create a new column to keep track of on-time/late deliveries
#there appears a warning due to pandas not knowing what it has to modifiy. If someone knows how to fix it do it please. Create a copy or whatever
df1['END'] = pd.to_datetime(df1['END'])
df1['DELIVERED'] = pd.to_datetime(df1['DELIVERED'])

df1['ON_TIME'] = (df1['END'] >= df1['DELIVERED']).astype(int)
df1.head()
```

```python
#it is normal that we're missing a lot of names in df4 (as we're currently using a small dataset)
#the important thing is that there are no missing names in df2

#now, I asked chatGPT this:
#"I want to do is to create new columns (or just one) for each translator in translators_df indicating which source-target language task
#they can manage. In the datasheet df4, many translators have multiple rows, one for each language pair, and each row has a different hourly_rate cost. How do I
#incorporate all that information in a single row for every translator in translators_df? Do I create a column for every possible language pair, do I create a single
#column which contains a dictionary for every translator where keys are language pairs and values are hourly_rate cost, or is it better to do another thing?"

#and it told me that for pandas, the best thing is to create a new column for each language pair, so that's what we're gonna do :D
```

```python
df4['PAIR'] = df4['SOURCE_LANG'] + "_" + df4['TARGET_LANG']

pivot = df4.pivot_table(
    index='NAME',
    columns='PAIR',
    values='HOURLY_RATE',
    aggfunc='first')

pivot = pivot.reset_index()

translators_df = translators_df.merge(pivot, on='NAME', how='left')

translators_df.head()
```
